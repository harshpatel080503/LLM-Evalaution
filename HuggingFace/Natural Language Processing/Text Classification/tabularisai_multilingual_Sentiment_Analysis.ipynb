## Multilingual Sentiment Classification Model (23 Languages)

Model Details :- 

Model Name: tabularisai/multilingual-sentiment-analysis

Base Model: distilbert/distilbert-base-multilingual-cased

Task: Text Classification (Sentiment Analysis)

Languages:<br>
Supports English plus Chinese (中文), Spanish (Español), Hindi (हिन्दी), Arabic (العربية), Bengali (বাংলা), Portuguese (Português), Russian (Русский), Japanese (日本語), German (Deutsch), Malay (Bahasa Melayu), Telugu (తెలుగు), Vietnamese (Tiếng Việt), Korean (한국어), French (Français), Turkish (Türkçe), Italian (Italiano), Polish (Polski), Ukrainian (Українська), Tagalog, Dutch (Nederlands), Swiss German (Schweizerdeutsch), and Swahili.

Number of Classes: 5 (Very Negative, Negative, Neutral, Positive, Very Positive)

Usage:

Social media analysis<br>
Customer feedback analysis<br>
Product reviews classification<br>
Brand monitoring<br>
Market research<br>
Customer service optimization<br>
Competitive intelligence<br>

===================================================

Model Description

This model is a fine-tuned version of distilbert/distilbert-base-multilingual-cased for multilingual sentiment analysis. It leverages synthetic data from multiple sources to achieve robust performance across different languages and cultural contexts.

Training Data

Trained exclusively on synthetic multilingual data generated by advanced LLMs, ensuring wide coverage of sentiment expressions from various languages.

Training Procedure

Fine-tuned for 3.5 epochs.
Achieved a train_acc_off_by_one of approximately 0.93 on the validation dataset.

Ideal for:

Multilingual social media monitoring
International customer feedback analysis
Global product review sentiment classification
Worldwide brand sentiment tracking

```python
import os
from huggingface_hub import InferenceClient
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
client = InferenceClient(
    provider = "hf-inference",
    api_key = os.environ["HF_TOKEN"]
)
```

```python
result = client.text_classification(
    """I wouldn’t say the policy was a disaster, but calling it a success feels dishonest. 
Some people benefited, others paid the price, and in the end nothing really changed.
""",
    model = "tabularisai/multilingual-sentiment-analysis"
)
```

```python
result
```

```python
result = client.text_classification(
    """Sure, it technically works, if your definition of success includes frustration, delays, and lowered expectations.
""",
    model = "tabularisai/multilingual-sentiment-analysis"
)
```
